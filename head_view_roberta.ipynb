```python
from bertviz.pytorch_transformers_attn import RobertaModel, RobertaTokenizer
from bertviz.head_view_bert import show
```

```python
%%javascript
require.config({
  paths: {
      d3: '//cdnjs.cloudflare.com/ajax/libs/d3/3.4.8/d3.min',
      jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
  }
});
```

```python
roberta_version = 'roberta-base'
# roberta_version = 'roberta-large'
# roberta_version = 'roberta-large-mnli'
model = RobertaModel.from_pretrained(roberta_version)
tokenizer = RobertaTokenizer.from_pretrained(roberta_version)
sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
show(model, tokenizer, sentence_a, sentence_b, bert_type='roberta')
```
