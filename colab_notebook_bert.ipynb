```python
# Import package functions

from bertviz.pytorch_transformers_attn import BertModel, BertTokenizer
from bertviz.head_view_bert import show as show_head
from bertviz.model_view_bert import show as show_model
from bertviz.neuron_view_bert import show as show_neuron
```

```python
# This is needed to use requirejs on google colab

def configure_plotly_browser_state_head():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          require.config({
            paths: {
              d3: '//cdnjs.cloudflare.com/ajax/libs/d3/3.4.8/d3.min',
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
            }
          });
        </script>
        '''))
  

def configure_plotly_browser_state_model():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          require.config({
            paths: {
              d3: '//cdnjs.cloudflare.com/ajax/libs/d3/5.7.0/d3.min',
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
            }
          });
        </script>
        '''))


def configure_plotly_browser_state_neuron():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          require.config({
            paths: {
              d3: '//cdnjs.cloudflare.com/ajax/libs/d3/5.7.0/d3.min',
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
            }
          });
        </script>
        '''))
```

```python
# Load the model

bert_version = 'bert-base-uncased'
do_lower_case = True
model = BertModel.from_pretrained(bert_version)
tokenizer = BertTokenizer.from_pretrained(bert_version, do_lower_case=do_lower_case)
```

```python
# Head view

configure_plotly_browser_state_head()

sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
show_head(model, tokenizer, sentence_a, sentence_b)
```

```python
# Model view

configure_plotly_browser_state_model()

sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
show_model(model, tokenizer, sentence_a, sentence_b)
```

```python
# Neuron view

configure_plotly_browser_state_neuron()

sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
show_neuron(model, tokenizer, sentence_a, sentence_b)
```
